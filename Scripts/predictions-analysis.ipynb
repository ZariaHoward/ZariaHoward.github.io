```python
# count top 5
c = Counter()
total = 0
for pred in data:
    for conf, i in sorted(zip(pred, np.arange(len(pred))), reverse=True)[:5]:
        c[i] += 1
        total += 1
```

```python
for label, count in c.most_common():
    print '{}% ({})'.format((100 * count) / total, count), labels[label]
```

```python
# count top 1
c = Counter()
total = 0
for pred in data:
    label = np.argmax(pred)
    c[label] += 1
    total += 1
```

```python
for label, count in c.most_common():
    print '{}% ({})'.format((100 * count) / total, count), labels[label]
```

```python
# count top 1 if it's greater than 0.9
c = Counter()
total = 0
for pred in data:
    if np.max(pred) > 0.9:
        label = np.argmax(pred)
        c[label] += 1
        total += 1
```

```python
for label, count in c.most_common():
    print '{}% ({})'.format((100 * count) / total, count), labels[label]
```
